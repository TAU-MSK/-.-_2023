## Данные для проекта:

   Технологический процесс:

- Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.

    Стадии процесса: 

1. Флотация

Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).

2. Очистка 

Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

    Описание данных:

условные обозначения технологического процесса:

    Rougher feed — исходное сырье
    Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
    Xanthate — ксантогенат (промотер, или активатор флотации);
    Sulphate — сульфат (на данном производстве сульфид натрия);
    Depressant — депрессант (силикат натрия).
    Rougher process (англ. «грубый процесс») — флотация
    Rougher tails — отвальные хвосты
    Float banks — флотационная установка
    Cleaner process — очистка
    Rougher Au — черновой концентрат золота
    Final Au — финальный концентрат золота
    
параметры этапов:

    air amount — объём воздуха
    fluid levels — уровень жидкости
    feed size — размер гранул сырья
    feed rate — скорость подачи
    
наименование признаков:
наименование признаков должно быть такое: rougher.input.feed_ag (пример);

возможные значения для блока[этап]:

    rougher — флотация
    primary_cleaner — первичная очистка
    secondary_cleaner — вторичная очистка
    final — финальные характеристики
    
возможные значения для блока [тип_параметра]:

    input — параметры сырья
    output — параметры продукта
    state — параметры, характеризующие текущее состояние этапа
    calculation — расчётные характеристики

#### Проверка правильности рассчёта эффективности обогащения 

Проверим, что эффективность обогащения рассчитана правильно:

    1. Вычислим эффективность обогащения на обучающей выборке для признака rougher.output.recovery;
    2. Найдём MAE между данным расчётом и значением признака;
    3. Опишем выводы.

Вычислим эффективность обогащения на обучающей выборке для признака rougher.output.recovery

```python
C = data_train['rougher.output.concentrate_au']
F = data_train['rougher.input.feed_au']
T = data_train['rougher.output.tail_au']
dressing_efficiency = ((C * (F - T)) / (F * (C - T))) * 100
dressing_efficiency
# вычисления производим по формуле для расчета эффективности обогащения золота, где:
#    C - доля золота в концентрате после флотации/очистки,
#    F - доля золота в сырье до флотации/очистки,
#    T - доля золота в отвальных хвостах после флотации/очистки.
  
```

Найдём MAE между данным расчётом и значением признака:

```python
mae = mean_absolute_error(data_train['rougher.output.recovery'], dressing_efficiency)
f"Средняя абсолютная ошибка (MAE): {mae}"
```

Вывод: малое значение MAE указывает на высокую точность расчетов эффективности обогащения в предоставленных данных

### Напишем функцию для вычисления итоговой sMAPE

```python
def smape(y_true, y_pred):
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2 + 1e-10
    return np.mean(np.abs(y_true - y_pred) / denominator) * 100

def final_smape(smape_rougher, smape_final):
    return 0.25 * smape_rougher + 0.75 * smape_final
```

### Обучим и выберем лучшую модель

Обучим разные модели и оценим их качество. Выберем лучшую модель и проверим её на тестовой выборке, далее сделаем выводы.

```python
models = {
    'LinearRegression': LinearRegression(),
    'DecisionTreeRegressor': DecisionTreeRegressor(random_state=42),
    'RandomForestRegressor': RandomForestRegressor(random_state=42)
}
# создадим словарь выбранных моделей

results = {}
# создадим словарь для хранения результатов sMAPE каждой модели

X_train_split, X_valid, y_rougher_train_split, y_rougher_valid = train_test_split(X_train, y_train_rougher, test_size=0.25, random_state=42)
_, _, y_final_train_split, y_final_valid = train_test_split(X_train, y_train_final, test_size=0.25, random_state=42)
# разделим обучающие данные на обучающую и валидационную выборки

for name, model in models.items():
    model.fit(X_train_split, y_rougher_train_split)
    smape_rougher = smape(y_rougher_valid, model.predict(X_valid))
    model.fit(X_train_split, y_final_train_split)
    smape_final = smape(y_final_valid, model.predict(X_valid))
    results[name] = final_smape(smape_rougher, smape_final)
# проведём обучение моделей и расчет sMAPE

for model_name, smape_score in results.items():
    print(f"{model_name}: Итоговая sMAPE = {smape_score:.2f}%")
# выведем результаты итоговой sMAPE
```

Вывод:  результаты обучения моделей LinearRegression, DecisionTreeRegressor и RandomForestRegressor показали, что из трех рассмотренных моделей RandomForestRegressor продемонстрировала наилучшую производительность с наименьшим значением итоговой sMAPE (4.65%).
